```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

# 直角坐标系验证马吕斯定律

```python
theta = np.linspace(90,0,7)
theta = (theta/180)*np.pi
I = np.array([0,18.9,67.9,137.1,202.3,246.1,260.4])
cos2 = np.cos(theta)**2
print(I+1.9)
print(I)
print(cos2)
print(I*cos2)
```

```python
plt.figure(figsize=(10,6))
plt.title(r"cos$\theta$**2与I的关系")
plt.ylabel(r"I")
plt.xlabel(r"cos$\theta$**2")
plt.grid()
plt.plot(cos2, I)
for a, b in zip(cos2.round(2), I.round(2)):  
    plt.text(a, b, (a,b),ha='center', va='bottom', fontsize=10) 

plt.savefig("光的偏振特性研究.jpg",dpi=150)
```

# 极坐标系分析圆、椭圆、线偏振光分布规律

```python
def draw(theta, I, name):
    plt.figure(figsize=(6,6))
    plt.figtext(0.52, 0.95, name, ha='center', size=10) # 标题
    ax = plt.subplot(111, projection='polar') # polar为极坐标
    ax.set_thetagrids(np.arange(0.0, 360.0, 30.0)) # 极角标注
    ax.plot(theta,I,lw=2,color="r")
    ax.plot(theta[[0,-1]],I[[0,-1]],lw=2,color="r")
    ax.grid(True) # 网格线
    plt.savefig(name.replace("$","").replace('\\',"")+'.jpg',dpi=150) #保存
```

```python
# 线偏振光
theta2 = np.array([338,8,38,68,98,128,158,188,218,248,278,308])
I2 = np.array([1.9,39.4,114.2,149.8,115.5,39.8,1.9,38.9,113.5,149.5,113.8,39.8])
theta2 = (theta2/180)*np.pi
I2 = I2 - 1.9
print(I2)
# 圆偏振光
theta3 = np.array([335,5,35,65,95,125,155,185,215,245,275,305])
theta3 = (theta3/180)*np.pi
I3 = np.array([63.5,66.7,83.6,87.1,78.1,66.3,60.1,68.5,79.8,88.2,84,66.3])
I3 = I3 - 1.9
print(I3)
# 椭圆偏振光
theta4 = np.array([5,35,65,95,125,155,185,215,245,275,305,335])
I4 = np.array([23.1,66.9,118.5,125.5,85.1,31.9,22.2,66.5,119.7,121.1,81.8,31.2])
theta4 = (theta4/180)*np.pi
I4 = I4 - 1.9
print(I4)
```

```python
draw(theta2, I2, r"线偏振光$\theta$与I的极坐标关系图")
```

```python
draw(theta3, I3, r"圆偏振光$\theta$与I的极坐标关系图")
```

```python
draw(theta4, I4, r"椭圆偏振光$\theta$与I的极坐标关系图")
```

# 根据布儒斯特角测定介质的折射率

```python
phi1 = np.array([277.9,14.0,30.8,60.5,14.9])
phi1_ = np.array([97.9,194.0,210.8,240.1,194.9])
phi2 = np.array([220.9,71.5,87.9,117.4,73.1])
phi2_ = np.array([40.9,251.5,267.5,297.0,253.1])
phi_p = (np.absolute(phi1-phi2)+np.absolute(phi1_-phi2_))/2
print(phi_p)
ave = np.average(phi_p)
print(np.tan((ave/180)*np.pi))
```

```python
np.sqrt(np.sum((phi_p-ave)**2)/20)
```
